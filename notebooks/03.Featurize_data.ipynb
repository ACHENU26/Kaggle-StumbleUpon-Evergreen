## Load datasets from CSV

```python
from mlxtend.utils.data import load_data
train_raw = load_data(name="train_raw")
val_raw = load_data(name="val_raw")
X_test_raw = load_data(name="test_inputs_raw")
train_raw = train_raw.drop(columns=['boilerplate'])
val_raw = val_raw.drop(columns=['boilerplate'])
X_test_raw = X_test_raw.drop(columns=['boilerplate'])
```

## Separate inputs and outputs

```python
from mlxtend.utils.data import df2Xydf
target = "label"
X_train_raw, y_train = df2Xydf(train_raw, target_column=target)
X_val_raw, y_val = df2Xydf(val_raw, target_column=target)
```

## Featurize inputs

```python
#import sys
#sys.path.append('../..')
#from gmsc.src.featurizer import featurization_fn
X_train = X_train_raw
X_val = X_val_raw
X_test = X_test_raw
```

## Concatenate inputs and outputs

```python
from pandas import concat
train = concat([X_train, train_raw[[target]]], axis=1)
val = concat([X_val, val_raw[[target]]], axis=1)
```

## Save featurized datasets to CSV

```python
from mlxtend.utils.data import save_data
save_data(train, name="train")
save_data(val, name="val")
save_data(X_test, name="test_inputs")
```
