# Split Data
___

## Split train_full_raw into train_raw and val_raw


```python
# Validation Size 
VAL_SIZE = 0.25
```

## Load raw train full

```python
from mlxtend.utils.data import load_data
train_full_raw = load_data(name="train_full_raw")
```

```python
train_full_raw.columns
```

## Split

```python
from sklearn.model_selection import train_test_split
train_raw, val_raw = train_test_split(train_full_raw, test_size=VAL_SIZE, random_state=0)
```

Save raw train and val

```python
from mlxtend.utils.data import save_data
save_data(train_raw, name="train_raw")
save_data(val_raw, name="val_raw")
```
