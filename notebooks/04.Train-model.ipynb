## Load train data

```python
from mlxtend.utils.data import *
train = load_data(name="train")
X_train, y_train = df2Xydf(train)

y_train
```

## Initialize model/pipeline and set hyper-parameters

```python
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler
from dabl import EasyPreprocessor
from sklearn.pipeline import make_pipeline

model = make_pipeline(EasyPreprocessor(), SGDClassifier(max_iter=1000, tol=1e-3))
```

## Fit model to train data

```python
model.fit(X_train, y_train)
```
