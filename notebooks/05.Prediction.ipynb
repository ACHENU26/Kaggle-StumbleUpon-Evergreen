## Load model and datasets

```python
from mlxtend.utils.data import load_model, load_data, save_data
model = load_model(name="model")
train = load_data(name="train")
val = load_data(name="val")
```

# PREDICT KAGGLE 


```python
from pandas import DataFrame, concat

test_inputs = load_data(name="test_inputs")
labels = model.predict(test_inputs)

test_inputs['label'] = labels

test_inputs = test_inputs[['urlid', 'label']]
test_inputs.columns

#save_data(test_inputs, name="kaggle_predict")
test_inputs.to_csv("../data/submission.csv", index=False)
```
