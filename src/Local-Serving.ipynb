# Local Model Serving


## Check dotenv file

This file is required by the app to find the model:

```bash
cat ../.env
```

## Start Flask

Make sure we're in the right environment (where flask and all ML libraries are installed):

```bash
conda activate full-stack-ml
```

Run Flask (the `&` sign makes it run in the background):

```bash
export FLASK_APP=src.local_app
flask run &
```

## Send request

```bash
curl --request POST --url http://localhost:5000/realtime \
--header 'Content-Type: application/json' \
--data '{
"RevolvingUtilizationOfUnsecuredLines": 0.11265426,
"age": 51,
"NumberOfTime30-59DaysPastDueNotWorse": 2,
"DebtRatio": 0.572492288,
"MonthlyIncome": 9400,
"NumberOfOpenCreditLinesAndLoans": 10,
"NumberOfTimes90DaysLate": 0,
"NumberRealEstateLoansOrLines": 1,
"NumberOfTime60-89DaysPastDueNotWorse": 0,
"NumberOfDependents": 2
}' | jq
```
